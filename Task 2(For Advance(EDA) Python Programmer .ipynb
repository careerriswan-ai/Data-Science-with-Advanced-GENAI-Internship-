Q. Maximum Product of Two Elements in an Array
Given the array of integers nums, you will choose two different indices i and j of that array. Return the maximum value of (nums[i]-1)*(nums[j]-1)

```python
class Solution(object):
    def maxProduct(self, nums):
        nums.sort()
        return (nums[-1] - 1) * (nums[-2] - 1)
```

```python
obj = Solution()
result = obj.maxProduct([3,4,5,2])
print(result)
```

Q.There are n soldiers standing in a line. Each soldier is assigned a unique rating value.

You have to form a team of 3 soldiers amongst them under the following rules:

Choose 3 soldiers with index (i, j, k) with rating (rating[i], rating[j], rating[k]).
A team is valid if: (rating[i] < rating[j] < rating[k]) or (rating[i] > rating[j] > rating[k]) where (0 <= i < j < k < n).

```python
class Solution(object):
    def numTeams(self, rating):
        n = len(rating)
        count = 0

        for j in range(n):
            left_small = left_large = 0
            right_small = right_large = 0

            for i in range(j):
                if rating[i] < rating[j]:
                    left_small += 1
                else:
                    left_large += 1

            for k in range(j + 1, n):
                if rating[k] > rating[j]:
                    right_large += 1
                else:
                    right_small += 1

            count += left_small * right_large
            count += left_large * right_small

        return count                            
        
```

```python
obj = Solution()
rating = [2,5,3,4,1]
print(obj.numTeams(rating))
```

Q.number-of-students-doing-homework-at-a-given-time
Given two integer arrays startTime and endTime and given an integer queryTime.

The ith student started doing their homework at the time startTime[i] and finished it at time endTime[i].

Return the number of students doing their homework at time queryTime. More formally, return the number of students where queryTime lays in the interval [startTime[i], endTime[i]] inclusive.

```python
class Solution(object):
    def busyStudent(self, startTime, endTime, queryTime):
        count = 0

        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1

        return count  
```

```python
obj = Solution()
print(obj.busyStudent([1,2,3], [3,2,7], 4))
```

Q.Given an integer num, return the number of steps to reduce it to zero.

In one step, if the current number is even, you have to divide it by 2, otherwise, you have to subtract 1 from it.

 

```python
class Solution(object):
    def numberOfSteps(self, num):
        steps = 0

        while num > 0:
            if num % 2 == 0:
                num = num // 2

            else:
                num = num - 1
            steps += 1

        return steps     
```

```python
obj = Solution()
print(obj.numberOfSteps(14))
```

Q.Counting Bits
Given an integer n, return an array ans of length n + 1 such that for each i (0 <= i <= n), ans[i] is the number of 1's in the binary representation of i

```python
class Solution(object):
    def countBits(self, n):
        ans = [0] * (n + 1)

        for i in range(1, n +1):
            ans[i] = ans[i // 2] + (i % 2)

        return ans  
```

```python
obj = Solution()
print(obj.countBits(5))
```
